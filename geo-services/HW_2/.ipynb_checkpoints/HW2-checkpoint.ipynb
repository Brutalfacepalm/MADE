```python
import folium
import random
import rtree

import geopandas as gpd
import numpy as np 

from OSMPythonTools.nominatim import Nominatim
from OSMPythonTools.overpass import overpassQueryBuilder, Overpass
from shapely.geometry import Point, LineString, Polygon
from haversine import haversine, Unit
```

```python
nominatim = Nominatim()
overpass = Overpass()

afghan_id = nominatim.query('Afghanistan').areaId()

query = overpassQueryBuilder(area=afghan_id, elementType=['relation'], selector='"name:UN:en"="Afghanistan"', includeGeometry=True)
result = overpass.query(query)
```

```python
afghan = result.elements()[0]

afghan_border = np.array(afghan.geometry()['coordinates'][0])[:,::-1]
afghan_points = gpd.points_from_xy(afghan_border[:,0], afghan_border[:,1])
afghan_polygon = Polygon(afghan_border)

[min_lat_bbox, min_lng_bbox, max_lat_bbox, max_lng_bbox] = afghan_polygon.bounds
height_bbox = abs(abs(max_lat_bbox) - abs(min_lat_bbox))
width_bbox = abs(abs(max_lng_bbox) - abs(min_lng_bbox))
```

```python
def gen_point(min_lat, min_lng, max_lat, max_lng):
    lat = random.uniform(min_lat, max_lat)
    lng = random.uniform(min_lng, max_lng)
    
    return lat, lng
```

```python
map_afghan = folium.Map(location=[afghan_polygon.centroid.x, afghan_polygon.centroid.y], 
                        zoom_start=1.5, 
                        tiles = 'cartodbpositron')

afghan_border_group = folium.FeatureGroup(name='afghanistan border')
steps_of_search_point_group = folium.FeatureGroup(name='steps of search point')
point_within_afghanistan_group = folium.FeatureGroup(name='point within afghanistan')
search_bounds_group = folium.FeatureGroup(name='search bounds')
point_outside_afghanistan_group = folium.FeatureGroup(name='point outside afghanistan')

folium.Polygon(
    afghan_border,
    name='afghanistan border',
    color='red',
    opacity=0.5,
    fill=True,
    fill_color='red',
).add_to(afghan_border_group)

        
folium.CircleMarker(
    location=[(min_lat_bbox + max_lat_bbox) / 2, (min_lng_bbox + max_lng_bbox) / 2],
    name='afghanistan border',
    radius=2,
    popup=f'Latitude: {round(afghan_polygon.centroid.x, 5)}, Longitude: {round(afghan_polygon.centroid.y, 5)}',
    color='blue',
    fill=True,
    fill_color='blue',

).add_to(point_outside_afghanistan_group)

min_lat, min_lng, max_lat, max_lng = -90, -180, 90, 180

prev_point = []

while True:
    lat, lng = gen_point(min_lat, min_lng, max_lat, max_lng)
    
    if prev_point:
        folium.Polygon(
        [prev_point,[lat, lng]],
        name='steps of search point',
        color='brown',
    ).add_to(steps_of_search_point_group)
    
    prev_point = [lat, lng]
    
    if Point([lat, lng]).within(afghan_polygon):
        folium.CircleMarker(
            location=[lat, lng],
            name='point within afghanistan',
            radius=5,
            popup=f'Latitude: {round(lat, 5)}, Longitude: {round(lng, 5)}',
            color='green',
            fill=True,
            fill_color='green',
        ).add_to(point_within_afghanistan_group)
        break
    else:
        lat_plus = abs(abs(lat) - abs(afghan_polygon.centroid.x))
        lng_plus = abs(abs(lng) - abs(afghan_polygon.centroid.y))
        
        min_lat, min_lng = afghan_polygon.centroid.x - lat_plus, afghan_polygon.centroid.y - lng_plus
        max_lat, max_lng = afghan_polygon.centroid.x + lat_plus, afghan_polygon.centroid.y + lng_plus
               
    
    folium.Polygon(
        [[min_lat, min_lng], 
         [max_lat, min_lng],
         [max_lat, max_lng],
         [min_lat, max_lng]],
        name='search bounds',
        color='#67E667',
        opacity=0.01,
        fill=True,
        fill_color='#67E667',
        fill_opacity=0.2,
    ).add_to(search_bounds_group)
        
        
    folium.CircleMarker(
        location=[lat, lng],
        name='points outside afghanistan',
        radius=2,
        popup=f'Latitude: {round(lat, 5)}, Longitude: {round(lng, 5)}',
        color='red',
        fill=True,
        fill_color='red',
    
    ).add_to(point_outside_afghanistan_group)


search_bounds_group.add_to(map_afghan)
afghan_border_group.add_to(map_afghan)
point_within_afghanistan_group.add_to(map_afghan)
point_outside_afghanistan_group.add_to(map_afghan)
steps_of_search_point_group.add_to(map_afghan)

folium.LayerControl().add_to(map_afghan)

map_afghan
```

```python
def get_name_capital(x):
    area_country = x[0]
       
    def _query(areaId, s):
        query = overpassQueryBuilder(area=areaId, elementType=['node'], selector=s, out='body')
        name_capital = overpass.query(query)
        return name_capital
    
    def _query_check_population(areaId, s):
        query = overpassQueryBuilder(area=areaId, elementType=['node'], selector=s, out='body')
        name_capital = overpass.query(query)
        return name_capital
    
    if int(_query_check_population(area_country, ['"place"="country"']).elements()[0].tags()['population'].replace(' ', '')) > 3*10**5:
        name_capital = _query(area_country, ['"capital"="yes"'])

        if not name_capital.countElements():
            name_capital = _query(area_country, ['"is_capital"="country"'])
            if not name_capital.countElements():
                name_capital = _query(area_country, ['"admin_level"="2"', '"place"="city"'])
                if not name_capital.countElements():
                    name_capital = _query(area_country, ['"capital"="3"'])


        name_capital_ru = name_capital.elements()[0].tags()['name:ru']
        name_capital = name_capital.elements()[0].tags()['name']
        
        return [name_capital_ru, name_capital]
    else:
        return [None, None]
```

```python
def get_capital_id(name_capital_ru, name_capital, area_country):
    
    def _query(areaId, s, elementType='relation', includeGeometry=True):
        query = overpassQueryBuilder(area=areaId, elementType=[elementType], selector=s, 
                                     includeGeometry=includeGeometry, out='body')
        res = overpass.query(query)
        return res
    
    def _query_search_capital(areaId, s):
        query = overpassQueryBuilder(area=areaId, elementType=['node'], selector=s, out='body')
        if overpass.query(query).elements() != []:
            return True
        return False
    
    
    def _query_search_capital_bbox(bbox, s):
        query = overpassQueryBuilder(bbox=bbox, elementType=['node'], selector=s, out='body')
        if overpass.query(query).elements() != []:
            return True
        return False
    
    def simple_query(s_capital, elementType='relation', includeGeometry=False):
        capital_id = _query(area_country, s_capital, elementType=elementType, includeGeometry=False).elements()
        return capital_id
    
    
    def non_simple_query(admin_level, elementType='relation', includeGeometry=True):
        s = [f'"admin_level"="{admin_level}"', '"boundary"="administrative"', '"type"="boundary"']
        areas = _query(area_country, s, elementType=elementType, includeGeometry=includeGeometry).elements()
        s_i = ['"capital"="yes"', '"is_capital"="country"']
        if areas != []:
            for areaId in areas:
                maybe_capital = _query_search_capital(3600000000 + areaId.id(), s_i)
                if maybe_capital:
                    return areaId.id()
    
    
    if name_capital_ru:
        if name_capital_ru == 'Дамаск':
            id_country = non_simple_query(admin_level = 4, includeGeometry=False)
            if id_country:
                return 3600000000 + id_country
                    
        if name_capital_ru == 'Дакка':
            id_country = non_simple_query(admin_level = 5, includeGeometry=False)
            if id_country:
                return 3600000000 + id_country
                    
        if name_capital_ru == 'Додома':
            s_capital = ['"place"="city"', '"is_capital"="country"', f'"name"="{name_capital}"']
            capital_id = simple_query(s_capital, elementType='way', includeGeometry=False)
            if capital_id:     
                return capital_id[0].id()
                               
        s_capital = ['"admin_level"="7"', f'"name:ru"="{name_capital_ru}"']
        capital_id = simple_query(s_capital, includeGeometry=False)
        if capital_id:     
            return 3600000000 + capital_id[0].id()
            
        s_capital = [f'"name:ru"="{name_capital_ru}"']
        capital_id = simple_query(s_capital, includeGeometry=False)
        if capital_id:     
            return 3600000000 + capital_id[0].id()
            
        s_capital = [f'"name"="{name_capital}"']
        capital_id = simple_query(s_capital, includeGeometry=False)
        if capital_id:     
            return 3600000000 + capital_id[0].id()
                                
        s = ['"admin_level"="7"', '"boundary"="administrative"', '"type"="boundary"']
        bboxes = _query(area_country, s).elements()
        if bboxes != []:
            s_bbox = ['"capital"="yes"']
            for bbox in bboxes:
                b = np.array(bbox.geometry()['coordinates'][0])[:,::-1]
                bbox_for_query = Polygon(b)
                bbox_for_query = bbox_for_query.bounds

                maybe_capital = _query_search_capital_bbox(bbox_for_query, s_bbox)
                if maybe_capital:
                    return 3600000000 + bbox.id() 
                    
                    
        id_country = non_simple_query(admin_level = 6)
        if id_country:
            return 3600000000 + id_country

        id_country = non_simple_query(admin_level = 2)
        if id_country:
            return 3600000000 + id_country
```

```python
def count_house(area_id):
    selectors = ['"building"="apartments"',
                '"building"="bungalow"',
                '"building"="cabin"',
                '"building"="detached"',
                '"building"="dormitory"',
                '"building"="farm"',
                '"building"="ger"',
                '"building"="hotel"',
                '"building"="house"',
                '"building"="houseboat"',
                '"building"="residential"',
                '"building"="semidetached_house"',
                '"building"="static_caravan"',
                '"building"="terrace"']
    counts = 0
    for type_object in ['node', 'way', 'relation']:
            for s in selectors:
                
                query = overpassQueryBuilder(area=area_id, elementType=[type_object], selector=s, out='count')
                result = overpass.query(query, timeout=600)
                
                counts += result.countElements()
    
    return counts
```

```python
def count_pharmacy(capital_id):
    if capital_id:
        query = overpassQueryBuilder(area=capital_id, elementType=['node'], selector='"amenity"="pharmacy"', out='count')
        result = overpass.query(query).countElements()
        return result
```

```python
def border(id_country, name_country):
    query = overpassQueryBuilder(area=id_country, elementType=['relation'], 
                                 selector=f'"name:ru"="{name_country}"', includeGeometry=True)
    result = overpass.query(query, timeout=100)
    
    s = []
        
    q = np.array(result.elements()[0].geometry()['coordinates'])
    if len(q)>1:
        for i in q:
            s += list(i[0])
    else:
        s = list(q[0])
    s.sort(key=lambda x:x[1])
    
    p = np.array([min(s, key=lambda x:x[1])]+[max(s, key=lambda x:x[1])])
    p_y1, p_y2, p_x1, p_x2 = p[0][1], p[1][1], p[0][0], p[1][0]
    k = (p_y1 - p_y2)/(p_x1 - p_x2)
    b = p_y1 - p_x1*k
    
    p1 = np.array([min(s, key=lambda x:x[0])]+[max(s, key=lambda x:x[0])])
    k1 = -1/k
    b1 = (p_y2 + p_y1)/2 - ((p_y2 + p_y1)/2 - b)/k*k1
    p1_x1, p1_x2 = p1[0][0], p1[1][0]
    p1_y1, p1_y2 = p1_x1*k1 + b1, p1_x2*k1 + b1  

    folium.Polygon(
        [[p1_y1, p1_x1],
         [p1_y2, p1_x2]],
        name='search bounds',
        color='black',
        opacity=0.2
    ).add_to(lines_border_south_group)
    return [k1, b1]
```

```python
def get_ratio_train_bus_stops(id_country, name_country):
    [k, b] = border(id_country, name_country)
    selectors_for_railway = ['"railway"="station"', '"railway"="halt"', '"railway"="platform"']
    selectors_for_bus = ['"highway"="bus_stop"', '"amenity"="bus_station"']
    
    eType = ['node', 'way']
    
    
    railway_stops = 0
    bus_stops = 0
    
    def cut(s, eType, incG=False):
        res = 0
        if eType != 'node':
            incG=True
        query = overpassQueryBuilder(area=id_country, elementType=[eType], 
                                     selector=s, out='body', includeGeometry=incG)
        stops = overpass.query(query, timeout=120).elements()  
        if stops:
            for stop in stops:
                if eType != 'node':
                    bord = []
                    g = np.array(stop.geometry()['coordinates'][0])
                    if len(g) > 1:
                        for i in np.array(stop.geometry()['coordinates']):
                            bord += list(i)
                    bord = np.array(bord).reshape((-1, 2))
                    if len(bord)>2:
                        poly = Polygon(bord)
                        x = poly.centroid.x
                        y = poly.centroid.y
                    else:
                        continue
                else:
                    y = stop.lat()
                    x = stop.lon()
                    
                if y <= k * x + b:
                    res += 1
            return res
    
    for et in eType:
        for s in selectors_for_railway:
            r = cut(s, et)
            if r:
                railway_stops += r

        for s in selectors_for_bus:
            r = cut(s, et)
            if r:
                bus_stops += r
        
    if bus_stops:
        ratio = railway_stops/bus_stops
    else:
        ratio = None
    
    return ratio
```

```python
def get_schools(s, eType):
    schools_query = overpassQueryBuilder(area=id_country_for_school, elementType=[eType], 
                                         selector=s, out='body', includeGeometry=True)
    return overpass.query(schools_query, timeout=120).elements()

schoolss = []
for et in ['way', 'relation']:
    for s in ['"amenity"="school"']:
        schls = get_schools(s, et)
        for schl in schls:
            schl_geo = schl.geometry()['coordinates']
            
            if len(schl_geo[0]) > 2:
                poly = Polygon(schl_geo[0])
        
                schoolss.append([poly.centroid.x, poly.centroid.y])

rtree_idx = rtree.index.Index()

for i_sch, school in enumerate(schoolss):
    rtree_idx.insert(i_sch, (school[0], school[1], school[0], school[1]))
    
nearest_schools = []
nn_rtree = list(rtree_idx.nearest((*point_in_afghanistan, *point_in_afghanistan), 5))

for indx in rtree_idx.intersection((-180, -90, 180, 90), objects=True):
    if indx.id in nn_rtree:
        nearest_schools.append([indx.bbox[1], indx.bbox[0]])
        
```
